```python
def cost(theta,X,Y):

    predictions = np.dot(X,theta.T)
    cost = ((predictions - Y)**2).mean(axis=0)
    return cost 
```

```python
def hypothesis(X, theta): 
    return np.dot(X, theta) 

def gradient(X, y, theta): 
    h = hypothesis(X, theta) 
    grad = np.dot(X.transpose(), (h - y)) 
    return grad

# function to compute the error for current values of theta 
def cost(X, y, theta): 
    h = hypothesis(X, theta) 
    J = np.dot((h - y).transpose(), (h - y)) 
    J /= 2
    return J


def batch_gradient_descent(X,Y,lr, theta ,IteraMax, N):

    theta_history= np.zeros((IteraMax,N+1))
    cost_history= np.zeros(IteraMax)
    error_list = []
    
    for it in range(IteraMax):
            
        grad = gradient(X, Y, theta)
        theta = theta - lr * gradient(X, Y, theta)
        #
        error_list.append(cost(X, Y, theta)) 
        #theta = theta - alpha*(X.T.dot((hypothese.flatten() - Y)))
        theta_history[it] = theta
        #cost_history[it] = cost

    return theta , theta_history, error_list
```
